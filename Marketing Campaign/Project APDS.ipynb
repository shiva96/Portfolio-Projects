## Identifying outliers 

---
We can use interquartile range to identify the outliers in the dataset. 
Setting the upper limit as 75% and lower limit as 25%

---

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x=df1['Income '])
plt.title('Box Plot of Income')
plt.show()
```

```python
#compute the IQR in the data set to find the normal range 

Q1 = df1['Income '].quantile(0.25)
Q3 = df1['Income '].quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
#finding the outliers in age column of the dataset

AQ1 = df1['age'].quantile(.25)
AQ3 = df1['age'].quantile(.75)
AIQR = AQ3- AQ1
AIQR
```

```python
#from the above we can see that there are some outliers in the Age column of our data set. 
Age_Lower_bound = AQ1 - 1.5*AIQR
Age_Upper_bound = AQ3 - 1.5*AIQR
print(Age_Lower_bound)
print(Age_Upper_bound)
```

```python
#now we will separate our outliers from the dataset into a different df column

df1_Filtered = df1[(df1['age'] >= Age_Lower_bound) & (df1['age'] <= Age_Upper_bound)]
df1_Filtered
df1_Filtered.isnull().sum()
```

```python
#Before we move ahead we need to remove any symbols from numeric columns like '$', ',' etc. 
# Step 1: Remove non-numeric characters from 'Income' column
df1_Filtered['Income '] = df1_Filtered['Income '].replace({'\$': '', ',': '', '\xa0': ''}, regex=True)

# Step 2: Convert the cleaned 'Income' column to numeric (float or int)
df1_Filtered['Income '] = pd.to_numeric(df1_Filtered['Income '], errors='coerce')

# Step 3: Handle any missing values (optional)
df1_Filtered.dropna(subset=['Income '], inplace=True)
```

## Hypothesis Testing

```python
import scipy.stats as stats

# Define the age threshold (e.g., 45 years old)
age_threshold = 45

# Group data based on age
older_individuals = df_one_hot[df_one_hot['age'] > age_threshold]
younger_individuals = df_one_hot[df_one_hot['age'] <= age_threshold]

# Compare their in-store shopping behavior
t_stat, p_val = stats.ttest_ind(older_individuals['NumStorePurchases'], younger_individuals['NumStorePurchases'])

print(f"T-statistic: {t_stat}, P-value: {p_val}")
if p_val < 0.05:
    print("Older individuals significantly differ in their in-store shopping preferences.")
else:
    print("No significant difference in in-store shopping preferences by age.")
```

```python
# Compare customers with and without children
customers_with_children = df_one_hot[df_one_hot['total_children'] > 0]
customers_without_children = df_one_hot[df_one_hot['total_children'] == 0]

# Compare their online shopping behavior
t_stat, p_val = stats.ttest_ind(customers_with_children['NumWebPurchases'], customers_without_children['NumWebPurchases'])

print(f"T-statistic: {t_stat}, P-value: {p_val}")
if p_val < 0.05:
    print("Customers with children significantly prefer online shopping.")
else:
    print("No significant difference in online shopping preference by the presence of children.")
```

```python
# Calculate correlation between store purchases and other channels
correlation_web_store = df_one_hot['NumWebPurchases'].corr(df['NumStorePurchases'])
correlation_catalog_store = df_one_hot['NumCatalogPurchases'].corr(df['NumStorePurchases'])

print(f"Correlation between Web and Store Purchases: {correlation_web_store}")
print(f"Correlation between Catalog and Store Purchases: {correlation_catalog_store}")

if correlation_web_store < 0:
    print("Potential market cannibalization between web and store sales.")
if correlation_catalog_store < 0:
    print("Potential market cannibalization between catalog and store sales.")
```

Note: Market cannibalization refers to a drop in sales and demand for a product when the company replaces it with a new one. Market cannibalization can occur when a new product is similar to an existing product, and both share the same customer base.

```python
df_one_hot['Country_US']
```

```python
# Separate US and non-US customers
us_customers = df_one_hot[df_one_hot['Country_US'] == 1]
non_us_customers = df_one_hot[df_one_hot['Country_US'] == 0]

# Compare total purchases
t_stat, p_val = stats.ttest_ind(us_customers['total_purchases'], non_us_customers['total_purchases'])

print(f"T-statistic: {t_stat}, P-value: {p_val}")
if p_val < 0.05:
    print("The US significantly outperforms the rest of the world in total purchase volumes.")
else:
    print("No significant difference between US and the rest of the world in total purchase volumes.")
```
